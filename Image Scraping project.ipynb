```python
#Importing libraries
import selenium
import pandas as pd
import time
from bs4 import BeautifulSoup

#Importing Selenium Webdriver
from selenium import webdriver

#Importing Required Exception 
from selenium.common.exceptions import StaleElementReferenceException, NoSuchElementException

#Importing requests
import requests

#importing regex
import re
```

```python
#Activating the Chrome Browser
driver=webdriver.Chrome('chromedriver.exe')
time.sleep(3)
```

```python
#Opening the homepage of Amazon.com
url='https://www.amazon.com/'
driver.get(url)

time.sleep(2)
```

```python
#Finding tag of Search bar and Search bttn  and giving 'Sarres' as input
search_bar = driver.find_element_by_id("twotabsearchtextbox")    # Locating searc_bar by id
search_bar.clear()                                               # clearing search_bar
search_bar.send_keys('sarees')                                   # sending sarees as  input to search bar
search_button = driver.find_element_by_xpath('//div[@class="nav-search-submit nav-sprite"]/span/input')       # Locating search_button by xpath
search_button.click()                                                      # Clicking the button to start search
```

```python
#Scraping  Saree image
Img_src=[]
start=0
end=6
for page in range(start,end):     #Loop for scraping 6 pages 
    url=driver.find_elements_by_xpath('//a[@class="a-link-normal s-no-outline"]')
    for i in url:
        Img_src.append(i.get_attribute("href"))
    next_bttn=driver.find_element_by_xpath('//ul[@class="a-pagination"]/li[4]/a')
    next_bttn.click()
    time.sleep(2)
    
    #Make Empty List
    img_sarees=[]
    
    
for url in Img_src:
    driver.get(url)
    time.sleep(2)
    
    try:
        
        img=driver.find_element_by_css_selector('#landingImage')
        source=img.get_attribute('src')
        img_sarees.append(source)
    except NoSuchElementException:
        
        img_sarees.append("Image Not Found")
        
        
for i in range(len(img_sarees)):
    print("Downloading......")
    response= requests.get(img_sarees[i])
    file=open(r"C:\Users\welcome\Desktop\python\Image Classification\sarees"+str(i)+".jpg","wb")
    file.write(response.content)
```
